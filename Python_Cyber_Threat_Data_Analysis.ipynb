```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("downloads/cybersecurity_threat_dataset.csv")

# Count frequency of attack types
attack_counts = df['attack_type'].value_counts()

print("Attack Type Frequency:\n", attack_counts)

# Bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=attack_counts.index, y=attack_counts.values, palette="viridis")
plt.title("Frequency of Cyber Attack Types")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(attack_counts.values, labels=attack_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Cyber Attack Types")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("downloads/cybersecurity_threat_dataset.csv")

# Group by attack type and severity
severity_attack = df.groupby(['attack_type', 'severity']).size().unstack(fill_value=0)

print("Severity by Attack Type:\n", severity_attack)

# Stacked bar chart
severity_attack.plot(kind='bar', stacked=True, figsize=(10,6), colormap="coolwarm")
plt.title("Severity Distribution Across Attack Types")
plt.xlabel("Attack Type")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.legend(title="Severity")
plt.show()

# Grouped bar chart (alternative view)
severity_attack.plot(kind='bar', figsize=(10,6), colormap="viridis")
plt.title("Severity Levels by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.legend(title="Severity")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("downloads/cybersecurity_threat_dataset.csv")

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Group by date (daily attacks)
daily_attacks = df.groupby(df['timestamp'].dt.date).size()

print("Daily Attack Counts:\n", daily_attacks.head())

# Line chart of daily attacks
plt.figure(figsize=(12,6))
daily_attacks.plot(kind='line', marker='o')
plt.title("Cyber Attack Trends Over Time (Daily)")
plt.xlabel("Date")
plt.ylabel("Number of Attacks")
plt.grid(True)
plt.show()

# Group by month (monthly attacks)
monthly_attacks = df.groupby(df['timestamp'].dt.to_period("M")).size()

# Bar chart of monthly attacks
plt.figure(figsize=(12,6))
monthly_attacks.plot(kind='bar', color="skyblue")
plt.title("Cyber Attack Trends Over Time (Monthly)")
plt.xlabel("Month")
plt.ylabel("Number of Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("downloads/cybersecurity_threat_dataset.csv")

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Group by month and severity
severity_trends = df.groupby([df['timestamp'].dt.to_period("M"), 'severity']).size().unstack(fill_value=0)

print("Severity Trends by Month:\n", severity_trends.head())

# Line chart for severity trends
severity_trends.plot(kind='line', marker='o', figsize=(12,6))
plt.title("Severity Trends of Cyber Attacks Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.legend(title="Severity Level")
plt.grid(True)
plt.show()

# Stacked area chart (alternative view)
severity_trends.plot(kind='area', alpha=0.6, figsize=(12,6))
plt.title("Severity Distribution Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.legend(title="Severity Level")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("downloads/cybersecurity_threat_dataset.csv")

# Define mapping for outcomes
def classify_outcome(status):
    if status == "Resolved":
        return "Blocked"
    else:  # Ongoing or Contained
        return "Successful"

df['attack_outcome'] = df['status'].apply(classify_outcome)

# Count outcomes
outcome_counts = df['attack_outcome'].value_counts()

print("Attack Outcome Frequency:\n", outcome_counts)

# Bar chart
plt.figure(figsize=(6,5))
sns.barplot(x=outcome_counts.index, y=outcome_counts.values, palette="Set2")
plt.title("Proportion of Successful vs Blocked Cyber Attacks")
plt.xlabel("Attack Outcome")
plt.ylabel("Count")
plt.show()

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(outcome_counts.values, labels=outcome_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Attack Outcome Distribution")
plt.show()
```
